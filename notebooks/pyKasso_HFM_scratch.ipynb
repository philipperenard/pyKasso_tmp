````python
#Correct distributions:
#if these do not add up to the correct amount, the closest approximation of the percentages will be calculated
#if [1], all inlets or outlets will be part of the same group 
#if [1,1,1...] all inlets/outlets will be iterated over individually
outlets_importance = [1,1]
inlets_per_outlet  = [1,4]
inlets_importance  = [2,3]
outlets_shuffle = False
inlets_shuffle  = False

#Shuffle outlet order
if outlets_shuffle == True:
    np.random.shuffle(catchment.outlets)
#Shuffle inlet order
if inlets_shuffle == True:
    np.random.shuffle(catchment.inlets)

def repartition_points(catchment, points, importance):
    '''Correct for integers in importance factors list not summing correctly to total number of points'''
    repartition = []
    proportion  = []
    total       = float(sum(importance))     #total number of points as assigned (this may not be the actual total)
    for i,n in enumerate(importance):        #get iteration index and number of points being assigned to that iteration
        proportion.append(float(n)/total)            #percent of outlets to use this iteration
        repartition.append(round(proportion[i]*len(points))) #number of outlets to use this iteration (need to round bc percentage will not result in whole number)
    repartition[-1] = len(points)-sum(repartition[0:-1])     #leftover outlets are assignd to last iteration
    return repartition

def distribute_outlets(catchment, repartition):
    '''Distribute points into iteration groups based on the corrected repartition from the importance settings'''
    outlets = []
    i = 0              #point counter
    for k,n in enumerate(repartition):        #get iteration index and number of points being assigned to that iteration
        for j in range(n):                    #loop over each point in current iteration
            outlets.append((catchment.outlets[i][0],catchment.outlets[i][1],k))   #append point with its iteration number
            i += 1                                                                #increment point index by 1
    return np.asarray(outlets)

def distribute_inlets(catchment, repartition, outlets):
    '''Distribute points into iteration groups based on the corrected repartition from the importance settings'''
    inlets = []
    i = 0              #point counter
    for k,n in enumerate(repartition):        #get iteration index and number of points being assigned to that iteration
        for j in range(n):                    #loop over each point in current iteration
            inlets.append((catchment.inlets[i][0],catchment.inlets[i][1],k,outlets[k,0], outlets[k,1]))   #append point with its iteration number
            i += 1                                                                #increment point index by 1
    return np.asarray(inlets)


#Assign outlets to iterations: 
repartition = repartition_points(catchment, catchment.outlets, outlets_importance) #correct for outlets_importance not summing to correct number of actual outlets
outlets = distribute_outlets(catchment, repartition)  #actually distribute outlets to iterations

#Assign inlets to outlets:
repartition = repartition_points(catchment, catchment.inlets, inlets_per_outlet)
inlets = distribute_inlets(catchment, repartition, outlets)
        
#Assign inlet iterations (within outlet iterations):
inlets_iterations = []
for o,outlet in enumerate(outlets):        #loop over outlets
    inlets_current = inlets[inlets[:,2]==o]  #select only inlets assigned to the current outlet
    repartition = repartition_points(catchment, inlets_current, inlets_importance)
    ##actually distribute inlets to iterations
    i = 0               #total inlet counter for current outlet
    for k,n in enumerate(repartition): #get number of inlets for each iteration    
        for j in range(n):                  #loop over each inlet in current iteration
            inlets_iterations.append((inlets_current[i,0], inlets_current[i,1], inlets_current[i,2], inlets_current[i,3], inlets_current[i,4], k))   #append inlet with its outlet number and inlet iteration number
            i += 1

print('outlets\n',outlets)
print('inlets\n', inlets)
#print(inlets_iterations)

inlets  = pd.DataFrame(inlets_iterations, columns=['x','y','outlet', 'outletx', 'outlety', 'inlet_iteration'])
outlets = pd.DataFrame(outlets,           columns=['x','y','outlet_iteration'])

inlets
````

### Inputs/outputs for fast-marching ###

For agd-hfm: <br>
```
hfmIn = Eikonal.dictIn({
    'model':             'Riemann3', #three-dimensional Riemannian eikonal equation (for 2D: Riemann2, for isotropic: Isotropic2 or 3)
    'order':              2,         #recommended setting: 2
    'exportValues':       1,         #output travel time field
    'exportGeodesicFlow': 1          #output walker paths
})
hfmIn.SetRect(sides=[[xmin,xmax],[ymin,ymax],[zmin,zmax]],dims=[xres,yres,zres])
for seed in springs:
    hfmIn['seed'] = seed
    for tip in tips:
        hfmIn['tips'] = [tip]
        hfmIn['metric']    =  F
        hfmOut = hfmIn.Run()
        paths = hfmOut['geodesics']
``` 
       
- seed - coordinates of travel time=0 location, i.e. the spring the conduits are traveling to (array of [x y z])
- tips - coordinates of starting points for walkers, i.e. the inlets (array of [[x1 y1 z1] [x2 y2 z2]]
- metric - travel cost through each cell, i.e. the inverse of the velocity (array of cost values, higher=slower to travel through). If using anisotropic FM, this will include travel cost both parallel and perpendicular to the cell's gradient.
    - to calculate travel cost metric, need:
        - DxV,DyV,DzV: x and y components of gradient in each cell of array, calculate from DEM or from geologic model orientations in each cell (arrays of dim (xres,yres)).
        - alpha: travel cost in same direction as gradient (array of dim (xres,yres) of values between 0 and 1, higher=slower)
        - beta: travel cost perpendicular to gradient (same format as alpha)
- model - which fast-marching algorithm to use (string, either Riemann2, Riemann3, Isotropic2, Isotropic3).
