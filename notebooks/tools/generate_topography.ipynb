```python
def generate_random_topography(nx, ny):
    return np.rint(np.random.rand(nx,ny))

def generate_slopy_topography(nx, ny):
    d = np.zeros((nx,ny))
    for y in range(ny):
        for x in range(nx):
            d[x][y] = x*y
    return d

def generate_valley_on_axis(nx, ny, zmax, zmin, axis=0):
    X = [0, int(nx/2), nx]
    Y = [zmax, zmin, zmax]
    d = np.zeros((nx,ny))
    for y in range(ny):
        for x in range(nx):
            if axis == 0:
                d[x][y] = np.interp(x, X, Y)
            else:
                d[x][y] = np.interp(y, X, Y)
    return d
```

```python
topo = generate_valley_on_axis(100,100,100,50, axis=0)
topo = generate_random_topography(100,100)
```
