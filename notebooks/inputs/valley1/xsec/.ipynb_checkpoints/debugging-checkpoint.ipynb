## Debugging xsec ###

```python
import importlib
import copy
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import agd
from agd import Eikonal
from agd.Metrics import Riemann
```

```python
xmin = 0;  xres = 100;  dx = 10.0;  xmax = xmin + xres*dx
ymin = 0;  yres = 80;   dy = 10.0;  ymax = ymin + yres*dy   
zmin = 0;  zres = 60;   dz = 10.0;  zmax = zmin + zres*dz
```

```python
inlets  = np.genfromtxt('inlets2D.txt')
outlets = np.asarray([np.genfromtxt('outlets2D.txt')])
geo     = np.genfromtxt('geo60x100.csv', delimiter=',')
dem     = np.ones((zres,xres))
cost    = np.zeros((zres,xres))
for iz in range(zres):
    for ix in range(xres):
        dem[iz,ix] = iz*3
        if geo[iz,ix] == 0:
            cost[iz,ix] = 0.99
        if geo[iz,ix] == 1:
            cost[iz,ix] = 0.2
        if geo[iz,ix] == 2:
            cost[iz,ix] = 0.8
        if iz <= 35:    
            #dem[iz,ix] = iz*0.5 + ix*1.5
            #dem[iz,ix] = dem[iz,ix] - (xres-ix)*(35-iz)*0.05
            dem[iz,ix] = 35*3
            
#print(inlets)
#print(outlets)
#print(geo.shape,  np.unique(geo))
#print(cost.shape, np.unique(cost))
print(dem.shape,  dem.min(), dem.max())
plt.imshow(dem, origin='lower', extent=[xmin,xmax, zmin,zmax], cmap='gray_r') #white=low, black=high
plt.scatter(inlets[:,0],  inlets[:,1])
plt.scatter(outlets[:,0], outlets[:,1])
```

```python
np.savetxt('dem60x100.csv',dem,delimiter=',', fmt='%.1f')
```

```python
seeds = []
for o,outlet in enumerate(outlets):
    seeds.append([xmax-outlet[0],outlet[1]])
seeds = np.asarray(seeds)
tips = []
for i,inlet in enumerate(inlets):
    tips.append([xmax-inlet[0],inlet[1]])
tips = np.asarray(tips)
```

```python
hfmIn = Eikonal.dictIn({
    'model':             'Riemann2', #three-dimensional Riemannian eikonal equation (for 2D: Riemann2, for isotropic: Isotropic2 or 3)
    'order':              2,         #recommended setting: 2
    'exportValues':       1,         #output travel time field
    'exportGeodesicFlow': 1          #output walker paths
})

hfmIn.SetRect(sides=[[xmin,xmax],[zmin,zmax]],dims=[xres,zres])
X2,Y2 = hfmIn.Grid()  #get model grid
print(hfmIn['dims'])

#Set the variables
#seeds = np.rot90(outlets)
#tips  = np.rot90(inlets)
hfmIn['seeds']  = seeds  #set the spring coordinates 
hfmIn['tips']   = tips     #set the inlet coordinates for current iteration 
Cratio = 2                       #ratio of travel cost parallel to gradient / perpendicular to gradient (higher Cratio means lower perpendicular travel cost i.e. lower contrast between parallel and perpendicular)
alpha = cost*dem
alpha = np.rot90(alpha)             #travel cost in same direction as gradient
beta  = alpha / Cratio   #travel cost perpendicular to gradient (should be higher than parallel cost) 
print(alpha.shape,beta.shape)
DxV,DzV = np.gradient(dem, dz,dx, axis=(0,1,))   #x and y components of gradient in each cell of array 
DxV = np.rot90(DxV); DzV = np.rot90(DzV)
print(DxV.shape,DzV.shape)
F = Riemann.needle([DxV,DzV], alpha,beta)  #convert gradient fields to hfm format & incorporate travel costs
#hfmIn['cost'] = alpha
hfmIn['metric'] = F  #set the travel cost through each cell

#Run
hfmOut = hfmIn.Run()
time   = hfmOut['values']
paths  = hfmOut['geodesics']

#Plot
f = plt.figure(figsize=(15,10))
f.add_subplot(131, aspect='equal')
plt.imshow(time, origin='lower', extent=[zmin,zmax,xmin,xmax], cmap='summer') #darker=faster
plt.xlabel('Travel time array ' + str(time.shape))
skip=(slice(None,None,5),slice(None,None,5))
q = plt.quiver(Y2[skip], X2[skip], DxV[skip], DzV[skip], scale=5)
f.add_subplot(132, aspect='equal')
plt.imshow(alpha, origin='lower', cmap='copper_r') #darker=faster
plt.xlabel('Cost array ' + str(alpha.shape))
f.add_subplot(133, aspect='equal')
plt.imshow(alpha, origin='lower', extent=[zmin,zmax,xmin,xmax], cmap='copper_r') #darker=faster
plt.xlabel('Travel paths')
plt.scatter(seeds[:,1],  seeds[:,0], c='orange')
plt.scatter(tips[:,1], tips[:,0])
for path in paths:                                  #loop over paths (one per tip/starting point)
    pathX = path[1,:]                               #get list of X coord
    pathY = path[0,:]
    plt.plot(pathX,pathY, c='c')
```
