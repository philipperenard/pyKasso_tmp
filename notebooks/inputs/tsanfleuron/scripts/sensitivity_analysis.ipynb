```python
import pykasso as pk
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

```python
catchment = pk.SKS('inputs/tsanfleuron598x789.yaml') 
catchment.set_parameter('fractures_mode', 'null') 
catchment.update_feature('fractures')
catchment.set_parameter('faults_mode', 'csv') 
catchment.update_feature('faults')
catchment.set_parameter('orientation_mode', 'surface')                        
catchment.set_parameter('algorithm', 'Riemann2')
catchment.compute_karst_network()
f = catchment.show()
```

```python
plt.rcParams.update({'font.family':'Arial'})
data=catchment.karst_simulations[-1]
d = [d.data for d in catchment.geology.data if d.data_key=="surface"][0]
d[d==-9999]=np.nan
d=np.transpose(data,(1,0))
df=np.flip(d)
fig,ax = plt.subplots(figsize=(10,5))
ax.set_aspect('equal')
im = plt.imshow(df, extent=catchment.grid.extent,origin='upper' )
plt.colorbar(label='altitude [m]')
plt.ylim(126800, 132050)   
```

### Inlets analysis


```python
inlets1 = pk.SKS('inputs/tsanfleuron598x789.yaml') # inlet set of the ones where tracer test were carried out
for i in range(5):
    inlets1.update_fractures()
    inlets1.compute_karst_network()
```

```python
inlets2 = pk.SKS('inputs/tsanfleuron598x789.yaml') # inlet set of mapped dollines
for i in range(5):
    inlets2.set_inlets_data('inputs/tsanfleuron_inlets_dollines.csv')
    inlets2.update_inlets()
    inlets2.update_fractures()
    inlets2.compute_karst_network()
```

```python
inlets2.compute_average_paths()
```

```python
inlets3 = pk.SKS('inputs/tsanfleuron598x789.yaml') # inlet set of both previous sets together
for i in range(5):
    inlets3.set_inlets_data('inputs/tsanfleuron_inlets_full.csv')
    inlets3.update_inlets()
    inlets3.update_fractures()
    inlets3.compute_karst_network()
```

```python
inlets3.compute_average_paths()
```

```python
inlets4 = pk.SKS('inputs/tsanfleuron598x789.yaml') # inlet set tracer test set + stochastic inlets
for i in range(5):
    inlets4.set_inlets_mode('composite')
    inlets4.set_inlets_number(10)
    inlets4.update_inlets()
    inlets4.update_fractures()
    inlets4.compute_karst_network()
```

```python
inlets4.compute_average_paths()
```
